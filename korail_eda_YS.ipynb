```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib import font_manager,rc
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothic'
ko2018 = pd.read_csv('C:\\dacon\\data\\KORAIL)플랫폼통합경진대회 데이터\\역별 승하차실적_2018.csv')
ko2019 = pd.read_csv('C:\\dacon\\data\\KORAIL)플랫폼통합경진대회 데이터\\역별 승하차실적_2019.csv')
ko2020 = pd.read_csv('C:\\dacon\\data\\KORAIL)플랫폼통합경진대회 데이터\\역별 승하차실적_2020.csv')
ko2021 = pd.read_csv('C:\\dacon\\data\\KORAIL)플랫폼통합경진대회 데이터\\역별 승하차실적_2021.csv')
```

```python

# sns.barplot(data = station18)

### 정차역별로 이용승객을 집계하기 위해 데이터프레임 분리
count18 = ko2018[['정차역','승차인원수','하차인원수','이용승객수']]
count19 = ko2019[['정차역','승차인원수','하차인원수','이용승객수']]
count20 = ko2020[['정차역','승차인원수','하차인원수','이용승객수']]
count21 = ko2021[['정차역','승차인원수','하차인원수','이용승객수']]
###정차역별로 이용승객수 집계하고 상위 10개 역 추리기
people18 = count18.groupby(by=['정차역'], as_index=False).sum().sort_values('이용승객수',ascending=False)[:10]
people19 = count19.groupby(by=['정차역'], as_index=False).sum().sort_values('이용승객수',ascending=False)[:10]
people20 = count20.groupby(by=['정차역'], as_index=False).sum().sort_values('이용승객수',ascending=False)[:10]
people21 = count21.groupby(by=['정차역'], as_index=False).sum().sort_values('이용승객수',ascending=False)[:10]

figure, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2)
figure.set_size_inches(18,12)

plot18 = sns.barplot(data = people18, x='정차역', y='이용승객수',ax=ax1)
plot18.set_title("2018년도 정차역별 이용승객수")
plot19 =sns.barplot(data = people19, x='정차역', y='이용승객수',ax=ax2)
plot19.set_title("2018년도 정차역별 이용승객수")
plot20 =sns.barplot(data = people20, x='정차역', y='이용승객수',ax=ax3)
plot20.set_title("2018년도 정차역별 이용승객수")
plot21 =sns.barplot(data = people21, x='정차역', y='이용승객수',ax=ax4)
plot21.set_title("2018년도 정차역별 이용승객수")

```
